```python
import pandas as pd
```

```python
df = pd.read_csv('./vgsales.csv')
df
```

### 1. ~*Which company is the most common video game publisher?*~

```python
df ['Publisher'].value_counts().idxmax() # return the row label of the maximum value
```

___

### 2. ~*What’s the most common platform?*~

```python
df ['Platform'].value_counts().idxmax()
```

___

### 3. ~*What about the most common genre?*~

```python
df ['Genre'].value_counts().idxmax()
```

___

### 4. ~*What are the top 20 highest grossing games?*~

```python
df.sort_values('Global_Sales', ascending=False).head(20)
```

___

## 5. ~*For North American video game sales, what’s the median?*~
    
###   5.1. ~Provide a secondary output showing ten games surrounding the median sales output~

###   5.2. ~assume that games with same median value are sorted in descending order~

```python
mn = df['NA_Sales'].median()

df[(df['NA_Sales'] <= mn+1) & (df['NA_Sales'] >= mn-1)].head(10)
# .sort_values('NA_Sales', ascending=False)
```

___

## 7. ~*The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?*~

```python
# wii avarage global sales
# avg = df[['Platform'].mean()
# avg
# other platforms avg global sales
w = df[df['Platform'] == 'Wii'].Global_Sales.mean()
p = df[df['Platform'] != 'Wii'].Global_Sales.mean()
print(f'({p}, {w})')
```
